# Plot generation

This notebook reads the result files and generates plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/home/rco/Desktop/projects/tfermion/TFermion/results/results.csv', delimiter  = ',', index_col = 'Methods')
```

```python
fig, ax = plt.subplots()

plt.ylabel("T-gate cost")
plt.yscale('log')
plt.xticks(ticks = range(len(df.index)), labels = df.index, rotation=90)

for col,color in zip(df.columns, ['gray','brown','m','red','darkorange', 'green','blue', 'black']):
    ax.scatter(df.index, df[col], label = col,facecolors="None", edgecolors=color, linewidth = 1)

ax.legend(bbox_to_anchor=(1, 1))

plt.show()
```
